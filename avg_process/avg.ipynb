```python
chapter = "{{{{剧情框|标题={title}|内容=\n{content}}}}}\n\n"
background = "{{{{剧情背景图|{{剧情背景调用|{bg}}}}}}}\n"
narrator = "{{{{对话旁白|text={text}}}}}\n"
dialogue_r = "{{{{对话右|仅文字={name}|text={text}}}}}\n"
dialogue = "{{{{剧情对话行|自定图片={pic}|name={name}|text={text}}}}}\n"
```

```python
import re
avg_text='../gfl-data-alpha/data/ch/asset/avgtxt/-52-{id}-1.txt'

full_text=''
titles=['里坎·狩兽序章','猎人巡地','靶向陷阱','薄暮迷宫','巫女营帐','心魔池沼','烽火连台','混沌中继','混沌中继']
for id in range(9):
    with open(avg_text.format(id=id),'r',encoding='utf-8') as f:
        # print(id,avg_text.format(id=id))
        content=''
        for line in f.readlines():
            line=line.strip()
            # print(line)
            speaker,other=line.split('||')
            sp=other.find(':')
            effect,text=other[:sp],other[sp+1:]
            if '<BIN>' in effect:
                bg=re.findall(r'<BIN>(.*)</BIN>',effect)[0]
                if bg!='9':
                    content+=(background.format(bg=bg))
            if not text:
                continue
            pic=re.findall(r'(.*)<Speaker>',speaker)[0]
            name=re.findall(r'<Speaker>(.*)</Speaker>',speaker)[0]
            text=re.sub(r'\+',r'<br>',text)
            for d in range(5,120,5):
                text=re.sub(f'<size={d}>(.*?)</size>',f'{{{{Size|{d*2//5}px|\g<1>}}}}',text)
            text=re.sub(r'<color=(.+?)>(.*?)</color>',r'{{Color|\g<1>|\g<2>}}',text)
            if not name:
                content+=(narrator.format(text=text))
            elif name=='指挥官':
                content+=(dialogue_r.format(name=name,text=text))
            else:
                content+=(dialogue.format(pic=pic,name=name,text=text))
    full_text+=chapter.format(title=titles[id],content=content)
print(full_text)
```
